### Importing Libraries

```python
from typing import Counter
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Read csv file and display all columns name

```python
calo = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2018/2018-09-04/fastfood_calories.csv")
print(calo.columns)
```

## Dataset | Data in tabular form

```python
calo
```

### Number of items served by the restaurants

```python
fig, ax=plt.subplots()

labels, values = zip(*Counter(calo['restaurant']).items())

indexes = np.arange(len(labels))
width = 1

colour=['yellow','maroon','skyblue','red','darkblue','orange','green','purple']

plt.bar(indexes, values, color=colour)
plt.xticks(indexes + width * 0, labels,rotation=45)

ax.set_xlabel("Restaurants")
ax.set_ylabel("Number of items in menu")
ax.set_title("Number of items served by the restaurants",loc='center')

plt.show()
```

Above presented bar plot depicts 8 different restaurants and their menu items. It is clearly seen that Taco Bell serves highest number of menu items while, Chick Fil - A serves the least amount of menu items to the customers.

### Amount of calories served by the restaurants

```python
fig, ax=plt.subplots()

plt.bar(calo["restaurant"],calo["calories"])
plt.xticks(indexes + width * 0, labels,rotation=45)

ax.set_xlabel("Restaurants")
ax.set_ylabel("Amount of Calories")
ax.set_title("Amount of calories served by the restaurants",loc='center')

plt.show()
```

Above presented bar plot illustrates amount of calories served by the various restaurants. As seen in the above plot Mcdonalds serves the highest number of calories and Taco Bell serves the least.
